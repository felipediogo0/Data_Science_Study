```python
# Exercício 1 - Crie uma função que imprima a sequência de números pares entre 1 e 20 (a função não recebe parâmetro) e 
# depois faça uma chamada à função para listar os números

#Fiz com parâmetro =)

def pair_number(y):

    x = 1 

    while x < y:
        if x%2 == 0:
            print(x)
        else:
            pass
        x += 1

    print("Finished!")

    
pair_number(10)    
```
